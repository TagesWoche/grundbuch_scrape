### search in kantonsblatt for grundbuch publications

```python
def parse_kantonsblatt(year, lastIssue):
    result = {}
    parseYear = int(year)
    for i in range(1, int(lastIssue)):
        issue = str(i).zfill(3)
        print(issue)
        for lookup in range(0, 30):
            url = 'https://www.kantonsblatt.ch/archive/article/' + year + '/' + issue + '/' + year + issue + '01' + str(lookup).zfill(3) + '.html'
            r  = requests.get(url)
            data = r.text
            soup = BeautifulSoup(data, 'html.parser')
            if soup.find(content="Grundbuch"):
                result[i] = [soup]
                result[i].append(soup.find('font').next.next.text)
                result[i].append(issue)
    df = pd.DataFrame.from_dict(result, orient='index')
    df = df.reset_index()
    df = df.drop('index', 1)
    df.columns = ['html', 'datum', 'issue']
    df.to_csv('grundbuch_scrape_' + year + '.csv')
    return df
```

```python
def extract_entries(location):
    result = {}
    for i, name in enumerate(location):
        nextIndex = 0
        if (nextIndex < len(location)):
                nextIndex = i + 1
                if(name.select('span.b')):
                    try:
                        result[i] = name.select('span.b')
                        result[i].append(location[nextIndex].text)
                        
                    except:
                        result[i] = ''

    return result
```

```python
def parse_grundbuch(dataframe, location):
    
    result = {}
    entries = {}
    #keys = ('location', 'datum', 'issue', 'grundbuch_raw')
    #entries = dict.fromkeys(keys)
    entries['location'] = location
    for index, row in dataframe.iterrows():
        myps = row['html'].findAll('p')
        entries['grundbuch_raw'] = ''
        for i, p in enumerate(myps):
            if (p.text.strip() == str(location)):
                selector = i + 1
                location_data = myps[selector].findAll("span", { "class" : "gr8" })
                entries['grundbuch_raw'] = extract_entries(location_data)
                
        entries['datum'] = row['datum']
        entries['issue'] = row['issue']
        #print(entries['datum'])
        result[index] = copy.copy(entries)
        
    dataframe = pd.DataFrame.from_dict(result, orient='index')
        
    return dataframe
```

```python
def create_dataframe(dataframe):
    
    df_result = pd.DataFrame(columns= ['street', 'entry', 'location', 'issue', 'datum', 'Sektionskennung', 'Eigentum bisher', 'Eigentum nun', 'Rest'] )
    
    for i in range(0, len(dataframe)):
        try:
            df = pd.DataFrame.from_dict(dataframe['grundbuch_raw'][i], orient='index')
            df = df.reset_index()
            df = df.drop('index', 1)
            df.columns = ['street', 'entry']
            df['street'] = df['street'].astype('str')
            df['street'] = df['street'].map(lambda x: x.lstrip('<span class="b">').rstrip(', </span>'))
            df['location'] = dataframe['location'][i]
            df['issue'] = dataframe['issue'][i]
            df['datum'] = dataframe['datum'][i]
            
            df = df.replace({'entry': {r'\r\n': ''}}, regex=True)
            df = df.replace({'entry': {r'\r': ''}}, regex=True)
            df = df.replace({'entry': {r'\\r': ''}}, regex=True)
            df = df.replace({'entry': {r'<br>': ''}}, regex=True)
            df = df.replace({'entry': {r'r>': ''}}, regex=True)
            df = df.replace({'entry': {r'</br': ''}}, regex=True)
            df = df.replace({'entry': {r'\s+': ' '}}, regex=True)
            
            
            df = df.replace({'street': {r'\r\n': ''}}, regex=True)
            df = df.replace({'street': {r'\r': ''}}, regex=True)
            df = df.replace({'street': {r'\\r': ''}}, regex=True)
            df = df.replace({'street': {r'r>': ''}}, regex=True)
            df = df.replace({'street': {r'</br': ''}}, regex=True)
            df = df.replace({'street': {r'<br': ''}}, regex=True)
            df = df.replace({'street': {r'<br>': ''}}, regex=True)

            df['entry'] = df['entry'].str.split('(?<!\d)(?<!St)(?<!J)(?<!M)(?<!\se)(?<!V)(?<!F)(?<!H)(?<!W)\.(?!\d)')
            for i in range(0, len(df['entry'])):
                try:
                    df.loc[i, 'Sektionskennung'] = df['entry'][i][0]
                except:
                    df.loc[i, 'Sektionskennung'] = ''     
                try:
                    df.loc[i, 'Eigentum bisher'] = df['entry'][i][1]
                except:
                    df.loc[i, 'Eigentum bisher'] = ''
                try:
                    df.loc[i, 'Eigentum nun'] = df['entry'][i][2]
                except:
                    df.loc[i, 'Eigentum nun'] = ''
                try:
                    df.loc[i, 'Rest'] = df['entry'][i][3]
                except:
                    df.loc[i, 'Rest'] = ''

            df = df.replace({'street': {r'\r\n': ''}}, regex=True)
            
            df = df.replace({'Eigentum bisher': {r'\r\n': ''}}, regex=True)
            df = df.replace({'Eigentum nun': {r'\r\n': ''}}, regex=True)
            df = df.replace({'Sektionskennung': {r'\r\n': ''}}, regex=True)
            df = df.replace({'Eigentum bisher': {r'Eigentum bisher: ': ''}}, regex=True)
            df = df.replace({'Eigentum bisher': {r'Eigentum  bisher: ': ''}}, regex=True)
            df = df.replace({'Eigentum bisher': {r'Eigentümer bisher: ': ''}}, regex=True)
            df = df.replace({'Eigentum bisher': {r'Eigentum: ': ''}}, regex=True)
            df = df.replace({'Eigentum nun': {r'Eigentum nun: ': ''}}, regex=True)
            df = df.replace({'Eigentum nun': {r'Eigentum  nun: ': ''}}, regex=True)
            df['Eigentum bisher'] = df['Eigentum bisher'].str.strip()
            df['street'] = df['street'].str.strip()
            df['Eigentum nun'] = df['Eigentum nun'].str.strip()

            df_result = df_result.append(df)
        
        except:
            continue
    df_result = df_result.reset_index()
    df_result = df_result.drop('index', 1)
    return df_result
    
```
